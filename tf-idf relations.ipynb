```python
import pandas as pd

questions = pd.read_csv('data/2018-12-10-Treatd/LemmatizedQA.csv', sep=';', header=0)
terms = pd.read_csv('data/2018-12-10-Treatd/LemmatizedTerms.csv', sep=";")
termset = set(terms['term'])
```

```python
rels = {}
for index, row in questions.iterrows():
    words = row['AnswerTerms']
    if type(words) is float:
        continue
    rels = getRelations(row['AnswerTerms'], rels)
    rels = getRelations(row['QuestionTerms'], rels)


    
            
            
```

```python
def getRelations(phrase, rels):
    questw = words.split()
    questw = [w for w in questw if w in termset]
    traversed = set()
    for w in questw:
        traversed.add(w)
        if w in termset:
            windices = [i for i,a in enumerate(questw) if a.lower() == w.lower()]
            wother = [i for i in questw if i.lower() not in traversed]
            wotraversed = set()
            #print(windices)
            for o in wother:
                if(o in wotraversed):
                    continue;
                woindices = [i for i,a in enumerate(questw) if a.lower() == o.lower()]
                wotraversed.add(o)
                #print(w, o, len(woindices))
                if len(woindices) == 0:
                    continue;
                else:
                    dist = float('inf')
                    for pos in woindices:
                        for opos in windices:
                            if abs(pos-opos)<dist:
                                dist = abs(pos-opos)
                    currel = rels.get(w+""+o)
                    oprel = rels.get(o+""+w)
                    if (currel is None) and (oprel is None):
                        rels.setdefault(w+""+o,[])
                        #First term
                        rels[w+""+o].append(w)
                        #Second term
                        rels[w+""+o].append(o)
                        rels[w+""+o].append(dist)
                        rels[w+""+o].append(len(windices))
                        print(len(windices))
                        rels[w+""+o].append(1)
                        continue;
                    elif not(oprel is None):
                        currel = oprel
                    meandist = (currel[2]+dist)/2
                    #Mean dist
                    currel[2]=meandist
                    #Total frequency
                    currel[3]=currel[3]+len(woindices)
                    #IDF
                    currel[4]=currel[4]+1
                    #print(rels[w+""+o])
    
    return rels
```

```python
#print(rels.get('vacaágua'))
df = pd.DataFrame(rels)
df = df.transpose()
#for index, row in df.iterrows():
#    for index2, row2 in df.iterrows():
#        if (row[0].strip()==row2[1].strip() and row[1].strip()==row2[0].strip()):
#            tdf.drop(index2, inplace=True)
#            break;

df.to_csv("data/2018-12-11-Treatd/TF-IDF-RAW-NODUPLICATE.csv")
```
