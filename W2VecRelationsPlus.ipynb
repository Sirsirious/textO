```python
import pandas as pd
import gensim
import logging
import spacy

nlp = spacy.load('pt')

corpus = pd.read_csv('data/Raw/quinhentas_perguntas.csv', sep=',', header=0)
answers = pd.read_csv('data/Raw/quinhentas_respostas.csv', sep=",", header=0)
corpus['resposta'] = answers['resposta']
```

```python
from spacy.lemmatizer import Lemmatizer
from spacy.lang.en import LEMMA_INDEX, LEMMA_EXC, LEMMA_RULES

lemmatizer = Lemmatizer(LEMMA_INDEX, LEMMA_EXC, LEMMA_RULES)

terms = dict()
ents = dict()
for i, row in corpus.iterrows():
    text = row['pergunta']+" "+row['resposta']
    doc = nlp(text)
    tokenized = [lemmatizer(token.orth_, token.pos_)[0] for token in doc if not (token.is_stop or token.is_punct)]
    entities = [ent.text for ent in doc.ents]
    print(entities)
    
    
    
    for term in tokenized:
        if terms.get(term) is None:
            terms[term] = 1
        else:
            terms[term] = terms.get(term)+1
    
    for entity in entities:
        if ents.get(entity) is None:
            ents[entity] = 1
        else:
            ents[entity] = ents.get(entity)+1
```

```python
def get_doc(qadf):
    """
    qadf: a pandas dataframe with one question and one answer per row.
    This function returns the doc statistics with term count.
    
    """
    doc_info = []
    for i, row in qadf.iterrows():
        text = row['pergunta']+" "+row['resposta']
        count = count_words(text)
        values = {'doc_id' : i, 'doc_length':count}
        doc_info.append(values)
    return doc_info
```

```python
def count_words(sent):
    """
    Return the total number of words in a sentence
    """
    nlp = spacy.load('pt')
    return len(nlp(sent))
    
```

```python
import spacy
from spacy.lemmatizer import Lemmatizer
from spacy.lang.en import LEMMA_INDEX, LEMMA_EXC, LEMMA_RULES
lemmatizer = Lemmatizer(LEMMA_INDEX, LEMMA_EXC, LEMMA_RULES)

def create_freq_dict(qadf):
    nlp = spacy.load('pt')
    freqDict_list = []
    for i, row in qadf.iterrows():
        text = row['pergunta']+" "+row['resposta']
        freq_dict = {}
        words = nlp(text)
        words = [token for token in words if not (token.is_punct or not token.is_alpha or token.is_stop)]
        for token in words:
            word = lemmatizer(token.orth_, token.pos_)[0]
            if word in freq_dict:
                freq_dict[word]+=1
            else:
                freq_dict[word]=1
            temp = {'doc_id' : i, 'freq_dict' : freq_dict}
        freqDict_list.append(temp)
    return freqDict_list
```

```python
def computeTF(doc_info, freqDict_list):
    TF_scores = []
    for tempDict in freqDict_list:
        docid = tempDict['doc_id']
        for k in tempDict['freq_dict']:
            temp = {'doc_id' : docid, 'TF_score' : tempDict['freq_dict'][k]/doc_info[docid-1]['doc_length'], 'key' : k}
            TF_scores.append(temp)
    return TF_scores
    
```

```python
import math
def computeIDF(doc_info, freqDict_list):
    IDF_scores=[]
    counter = 0
    for dic in freqDict_list:
        counter +=1
        for k in dic['freq_dict'].keys():
            count = sum([k in tempDict['freq_dict'] for tempDict in freqDict_list])
            temp = {'doc_id' : counter, 'IDF_score':math.log(len(doc_info)/count), 'key': k}
            
            IDF_scores.append(temp)
    return IDF_scores
```

```python
def computeTFIDF(TF_scores, IDF_scores):
    TFIDF_scores = []
    for j in IDF_scores:
        for i in TF_scores:
            if j['key']==i['key'] and j['doc_id']==i['doc_id']:
                temp = {'doc_id' : j['doc_id'],
                       'TFIDF_score' : j['IDF_score']*i['TF_score'],
                       'key': i['key']}
            else:
                continue
            TFIDF_scores.append(temp)
    return TFIDF_scores
```

```python
## Takes long
doc_info = get_doc(corpus)
```

```python
freqDict_list = create_freq_dict(corpus)
TF_scores = computeTF(doc_info, freqDict_list)
IDF_scores = computeIDF(doc_info, freqDict_list)
```

```python

TFIDF_scores = computeTFIDF(TF_scores, IDF_scores)
```

```python
sortedList = sorted(TFIDF_scores, key=lambda k: k['TFIDF_score'], reverse=True)
```

```python
termset = [entry['key'] for entry in sortedList[:2600]]
```

```python
logging.basicConfig(format='%(asctime)s: %(levelname)s : %(message)s', level=logging.INFO)
```

```python
import spacy
from spacy.lemmatizer import Lemmatizer
from spacy.lang.en import LEMMA_INDEX, LEMMA_EXC, LEMMA_RULES
lemmatizer = Lemmatizer(LEMMA_INDEX, LEMMA_EXC, LEMMA_RULES)

def corpusreader(text):
    nlp = spacy.load('pt')
    for i, row in text.iterrows():
        if(i%10 == 0):
            logging.info("Read {0} reviews".format(i))
        mylist = [lemmatizer(token.orth_, token.pos_)[0] for token in nlp(row['pergunta']+" "+row['resposta'])]
        yield [term for term in mylist if term in termset]

documents = list(corpusreader(corpus))
logging.info("Done reading data file")
```

```python
model = gensim.models.Word2Vec(documents, size = 5, window =10, min_count=2, workers=4)
model.train(documents, total_examples=len(documents), epochs=10)
```

```python
relations = pd.DataFrame()
for term in termset:
    if term in model.wv.vocab:
        for otherTerm in model.wv.most_similar(positive=term):
            print(otherTerm)
            relations = relations.append({'Term1':term, 'Term2':otherTerm[0]}, ignore_index=True)
            #print(term, otherTerm[0])
```
