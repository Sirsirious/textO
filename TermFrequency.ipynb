```python
infile = open('data/Relations500p500r.csv', 'r', encoding='latin-1')

for line in infile:
    terms = line.split(',')
    firstterm = terms[0]
    for raw in terms[1:]:
        raw = raw.split("[")
        term = raw[0]
        dist = raw[1].split("]")[0]
        print(firstterm+" -> "+term.strip()+" || Average distance: "+dist)
```
