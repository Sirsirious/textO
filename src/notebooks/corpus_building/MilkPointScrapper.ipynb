```python
import requests
import re
import math
import os
import datetime
import sys
import unidecode, unicodedata
from bs4 import BeautifulSoup

now = datetime.datetime.now()
system = sys.platform
bar = "\\" if system == "win32" else "/"
if not os.getcwd().split('\\')[-1]=='data':
    os.chdir("../../../data/")
    
```

```python
rootdirectory = os.getcwd()+bar+"MilkPoint"+bar+str(now)[:10]+bar
categorias = soup.find_all('h5', class_='forum-categoria-nome')
#Number of pages is arbitrary to escape pages that are not related to dairy farming
for i in range(1, 9):
    page = requests.get('https://www.milkpoint.com.br/forum/categoria/'+str(i))
    soup = BeautifulSoup(page.content, 'html.parser')
    
    category = soup.find(class_='forum-categoria-nome').get_text().strip().split('\n')[0].replace('\r','').strip().lower()
    category = unidecode.unidecode(category)
    
    responses = soup.select("h5.forum-categoria-nome span")[0].get_text()
    numresponses = responses[responses.find("(")+1:responses.find(")")]
    subpages = math.ceil(int(numresponses)/30)
    if not os.path.exists(rootdirectory+category):
        os.makedirs(rootdirectory+category)

    for j in range(1, subpages+1):
        page = requests.get("https://www.milkpoint.com.br/forum/categoria/"+str(i)+"/?pg="+str(j))
        soup = BeautifulSoup(page.content, 'html.parser')
        topics= soup.select('tr')
        for tag in topics:
            topic = tag.get('id').replace('tr','')
            topicpage = requests.get("https://www.milkpoint.com.br/forum/topico/"+topic)
            topicsoup = BeautifulSoup(topicpage.content, 'html.parser')
            topictitle = topicsoup.find(class_="forum-topico-titulo").get_text().split("(")[0].strip()
            topictext = topicsoup.find(class_="bdB-cinza item-forum-topico-texto").get_text().strip()
            filename = rootdirectory+category+bar+topic+'.txt'
            f = open(filename, "w")
            f.write(topictitle)
            f.write('\n')
            f.write(topictext)
            f.close()
            print(topic,"written successfully,", str(int(numresponses)-((j-1)*30)-j),"topics remaining in this category.")
```
