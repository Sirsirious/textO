```python
import pandas as pd
edges = pd.read_csv('data/2018-12-11-Treatd/edges.csv', sep=";")
tfidf = pd.read_csv('data/2018-12-11-Treatd/TF-IDF-TFIDF.csv')

edges['source']= edges['source'].str.strip()
edges['target']= edges['target'].str.strip()

edges.head()
```

```python
#tfidf.head()
tfidf.set_index('index', inplace=True)
```

```python
inverseedges = edges.copy()


inverseedges['index'] = inverseedges['target']+inverseedges['source']
edges['index']= edges['source']+edges['target']
inverseedges.set_index('index', inplace=True)
edges.set_index('index', inplace=True)
```

```python
final = tfidf.append(edges)
final = final.append(inverseedges)

final.head()
```

```python
filtered = final[~final.index.duplicated(keep=False)]
```

```python
final.describe()
```

```python
filtered.describe()
```

```python
print("Total duplicates removed: ",len(final.index)-len(filtered.index))
```

```python
filtered.to_csv('data/2018-12-12-Treatd/FilteredTFIDFFromMapped',sep=";")
```
